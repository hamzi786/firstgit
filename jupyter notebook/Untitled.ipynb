# Different Plots in Data visualization

## Barplots

# - No. 01

```python
import pandas as pd
chilla=pd.read_csv("data_chilla2.csv")

import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="ticks", color_codes=True)

sns.barplot(x="Age", y='How many hours you code a day? (int) e.g: 5,4,3', hue="Location", data=chilla, color="blue", palette="pastel", saturation=1, dodge=True,  )
plt.show()
```

## Countplots

# - No. 01

```python
import pandas as pd
from pandas.core.frame import DataFrame
chilla=pd.read_csv("data_chilla2.csv")

import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="ticks", color_codes=True)

sns.countplot(x="Gender", hue="Location", data=chilla, color="red", palette="pastel", saturation=1, dodge=True, )
plt.show()
```

# scatterplots

# -No. 01

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="ticks", color_codes=True)
titanic = sns.load_dataset("titanic")
g=sns.FacetGrid(titanic, row="sex", hue="alone")
g=(g.map(plt.scatter, "age", "fare").add_legend())
plt.show()
```

# Lineplots

# -No. 01

```python
import pandas as pd
chilla=pd.read_csv("data_chilla2.csv")

import seaborn as sns
sns.set_theme(style="ticks")

palette = sns.color_palette("rocket_r")

sns.relplot(
    data=chilla,
    x="Age", y="What are you",
    hue="Location", size="Gender", col="Marital Status?",
    kind="line", size_order=["Male", "Female"], palette="pastel",
    height=5, aspect=.75, facet_kws=dict(sharex=False),
)
```

# Plotly 3D PLots

# -No. 01

```python
import pandas as pd
chilla=pd.read_csv("data_chilla2.csv")

import plotly.express as px


fig = px.box(chilla, x="Location", y="Age", color="Gender")
fig.update_traces(quartilemethod="exclusive")
fig.show()
```

# -No. 03

```python
import plotly.graph_objects as go

import pandas as pd

z_data = pd.read_csv('data_chilla.csv')

fig = go.Figure(data=[go.Surface(z=z_data.values)])
fig.update_traces(contours_z=dict(show=True, usecolormap=True,
                                  highlightcolor="limegreen", project_z=True))
fig.update_layout(title='My Data of Chilla', autosize=False,
                  scene_camera_eye=dict(x=1.87, y=0.88, z=-0.64),
                  width=500, height=500,
                  margin=dict(l=65, r=50, b=65, t=90)
)

fig.show()
```
